Uses python3.8

```python
from optimus import Optimus

op = Optimus("pandas")

df = op.create.dataframe({
    'A': ['a', 'b', 'c', 'd'],
    'B': [1, 3, 5, 7],
    'C': [2, 4, 6, None],
    'D': ['1980/04/10', '1980/04/10', '1980/04/10', '1980/04/10'],
    'E': [1998, 1998, 2004, 2001],
})
display(df)
```

```python
df.cols.quality()
```

```python
df.profile()
```

```python
df.cols.infer_types("A")
```

```python
# df.cols.count_by_dtypes("*")

df.cols.__dict__

t_methods = [method_name for method_name in dir(df.cols)
                  if callable(getattr(df.cols, method_name))]
print(t_methods)
```
